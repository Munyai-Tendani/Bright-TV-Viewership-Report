```sql
select * 
from `workspace`.`brighttv`.`viewership` 
limit 100;

SELECT COALESCE(UserID0, 0) AS Profile_ID,
       `Gender`,
       `Race`,
--Age
       `Age`,
CASE 
    WHEN `Age` >= 30 THEN "1. Adult"
    WHEN `Age` BETWEEN 20 AND 29 THEN "2. Young Adult"
    WHEN `Age` BETWEEN 13 AND 19 THEN "3. Teenager"
    WHEN `Age` BETWEEN 0 AND 12 THEN "4. Kid"

END AS Age_Classification,
---------------------------------------------------------------------------------------
--Channels watched

       `Channel2` As Channel,
--------------------------------------------------------------------------------------
--Watch Time Engagement
       date_format(`Duration 2`,'HH:mm:ss') AS Time_Engagement, 
CASE
    WHEN date_format(`Duration 2`,'HH:mm:ss') >= '01:00:00' THEN 'Highly Engaged'
    WHEN date_format(`Duration 2`,'HH:mm:ss') > '00:30:00' THEN 'Moderately Engaged'
    WHEN date_format(`Duration 2`,'HH:mm:ss')  <= '00:29:59' THEN'Low Engaged'
END AS Engagement_Level,

-----------------------------------------------------------------------------------------
--Extract date from RecordDate2  

date_format(`RecordDate2`, 'yyyy-MM-dd') AS Watch_Date,
--------------------------------------------------------------------------------------------
--Time bucket
date_format(`RecordDate2` + INTERVAL 2 HOURS,'HH:mm:ss') AS Watch_SA_Time,
CASE 
    WHEN date_format(`RecordDate2` + INTERVAL 2 HOURS, 'HH:mm:ss') BETWEEN '00:00:00' AND '04:59:59' THEN '1. Mid-night'
    WHEN date_format(`RecordDate2` + INTERVAL 2 HOURS, 'HH:mm:ss') BETWEEN '05:00:00' AND '11:59:59' THEN '2. Morning'
    WHEN date_format(`RecordDate2` + INTERVAL 2 HOURS, 'HH:mm:ss') BETWEEN '12:00:00' AND '17:59:59' THEN '3. Afternoon'
    ELSE '4. Evening'
END AS Time_Bucket,
 
--Date
MONTHNAME(`RecordDate2`) As Month,
DAYNAME(`RecordDate2`) As Day,
----------------------------------------------------------------------------------

--Provice

`Province`

FROM `workspace`.`brighttv`.`user-profiles` AS A
FULL OUTER JOIN `workspace`.`brighttv`.`viewership`AS B
ON A.`UserID` = B.`UserID0`;
```
